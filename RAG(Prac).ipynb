```python
from langchain.chains import create_retrieval_chain
from langchain.prompts import ChatPromptTemplate
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from operator import itemgetter

# Function to format the documents for the prompt
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

# Define the prompt template for the AI to respond to the user's query
prompt_str = """
Answer the user question briefly:

Question: {question}

conversation_history: {chat_history}
"""
prompt = ChatPromptTemplate.from_template(prompt_str)

# Number of document chunks to fetch
num_chunks = 2

# Setting up the retriever for context
retriever = qdrant.as_retriever(search_type="similarity", search_kwargs={"k": num_chunks})

# Setup the HuggingFace model
llm1 = ChatHuggingFace(llm=HuggingFaceEndpoint(repo_id="HuggingFaceH4/zephyr-7b-beta"))

# Fetch the question and conversation history from input
query_fetcher = itemgetter("question")
history_fetcher = itemgetter("chat_history")

# Initialize conversation history to store questions and responses
history = []

# Set up the prompt and retrieval chain logic
setup = {"question": query_fetcher, "chat_history": history_fetcher}

# Define the retrieval chain to get a response based on the user's question and context
chain = setup | prompt | llm1

# Function to handle the conversation and keep track of history
def handle_conversation(query):
    global history

    # Add the user's question to the conversation history
    user_question = f"user_question: {query}"

    # Pass the current conversation history as context to the model
    context = "\n".join([f"{entry[0]}: {entry[1]}" for entry in history])

    # Invoke the retrieval chain with the current question and conversation history
    response = chain.invoke({"question": query, "chat_history": context})

    # Format the AI's response and append both the question and response to the history
    ai_response = f"ai_response: {response}"
    history.append((user_question, ai_response))

    # Return the AI's response and the updated conversation history
    return response, history

# Example of a conversational exchange
query = "Tell me about ai policy"

# Get the response and the updated conversation history
response, updated_history = handle_conversation(query)

print("AI Response:", response)
print("Updated Conversation History:")
for entry in updated_history:
    print(f"{entry[0]}\n{entry[1]}")
```
